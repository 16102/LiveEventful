# Birthday Event

```python
from diffusers import DiffusionPipeline
import matplotlib.pyplot as plt
import torch
import gradio as gr
pipeline = DiffusionPipeline.from_pretrained("anusha-bhambore/live-eventful")
pipeline = pipeline.to("cuda")
```

```python
def generate_image_interface(prompt, gender, negative_prompt,num_inference_steps=50, weight=640):
  params = {'prompt': prompt, 'num_inference_steps': num_inference_steps, 'num_images_per_prompt':2, 'height':int(1.2*weight),
            'weight': weight, 'negative_prompt': negative_prompt, 'gender': gender}

  img = pipeline(**params).images
  return img[0],img[1]
description = "Experience the magic of personalized birthday event design with our innovative web app! Simply input your preferences and prompts, and watch as your creative ideas transform into stunning, one-of-a-kind birthday event images. "
demo = gr.Interface(generate_image_interface,title="Birthday Events", inputs=["text","text", "text",gr.Slider(1,100),gr.Slider(512,640)], outputs=["image","image"], description=description)
```

```python
demo.launch(share=True)
```

```python
from diffusers import DiffusionPipeline
import matplotlib.pyplot as plt
import torch
import gradio as gr
pipeline1 = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0")
pipeline1 = pipeline1.to("cuda")
```

```python
def generate_image_interface(prompt, negative_prompt, no_of_attendees, num_inference_steps=50, weight=640):
  params = {'prompt': prompt, 'num_inference_steps': num_inference_steps, 'num_images_per_prompt':2, 'height':int(1.2*weight),
            'weight': weight, 'negative_prompt': negative_prompt, "no_of_attendees" : no_of_attendees}

  img = pipeline1(**params).images
  return img[0],img[1]
description = "Experience the magic of personalized Wedding event design with our innovative web app! Simply input your preferences and prompts, and watch as your creative ideas transform into stunning, one-of-a-kind wedding event images. "
demo2 = gr.Interface(generate_image_interface,title="Wedding images", inputs=["text","text", "text",gr.Slider(1,100),gr.Slider(512,640)], outputs=["image","image"], description=description)
```
